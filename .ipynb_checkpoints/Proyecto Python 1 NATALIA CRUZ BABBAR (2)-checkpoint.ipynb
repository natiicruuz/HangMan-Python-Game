Desarrolle un juego de "**Ahorcado**" que cumpla con las siguientes características:

Al iniciar el juego, debe consultársele al usuario su **nombre**, un **nivel de dificultad (libre, fácil, intermedio, difícil)** y el **número de letras** de la palabra que tendrá que descubrir. Por ejemplo, si el usuario solicita una palabra de 5 letras, el juego puede generar la palabra "``carro``" o "``robot``". Esta palabra **no será revelada al usuario hasta el final del juego**. Se le proveerá un archivo de palabras para usar como fuente (`palabras.txt`).

Una vez iniciado el juego, debe solicitársele una letra al usuario en cada turno. Si la letra se encuentra presente en la palabra, debe revelársele al usuario **todas las ubicaciones de esa letra dentro de la palabra**. Por ejemplo, si la palabra secreta es "``robot``" y el usuario adivina la ``b``, el programa debe mostrar algo similar a:

```
_ _ b _ _
```
Si luego de esto el usuario adivina la ``o``, el programa debe actualizar la palabra con las letras adivinadas anteriormente, además de **todas** las ubicaciones de la letra ``o``:

```
_ o b o _
```

Si el usuario elige una letra que no está presente en la palabra, debe indicársele que ha fallado. El usuario pierde el juego al alcanzar un número específico de fallos que depende del nivel de dificultad elegido:
* **libre**: Los fallos se cuentan pero no hay límite.
* **fácil**: El usuario pierde al alcanzar los 12 fallos.
* **intermedio**: El usuario pierde al alcanzar los 8 fallos.
* **difícil**: El usuario pierde al alcanzar los 5 fallos.

Finalmente, si el usuario elige una letra que ya había provisto antes (independientemente de si fue un fallo o un acierto), el programa debe indicárselo y pedirle otra letra sin ninguna consecuencia.

Una vez termine el juego, si el usuario adivinó la palabra, debe indicársele cuantos intentos tomó en llegar al resultado, cuántos fallos acumuló y mostrársele un mensaje de felicidades. Si el usuario alcanzó el límite de fallos, debe indicársele que ha perdido, y revelar la palabra completa. Finalmente, debe consultársele al jugador si desea jugar de nuevo. Si contesta positivamente, el juego reiniciará preguntándole nuevamente los datos de configuración del juego (el nivel de dificultad y el número de letras).

### Usando clear_output
La función `clear_output` permite borrar el output de la celda en la que se ejecuta.

````python
import random
from IPython.display import clear_output
from time import sleep

with open ("palabras.txt") as lista_palabras:
    lista_palabras = lista_palabras.read().split()
    

def datosUsuario(lista_palabras):
    print("=====================================================================")
    print("BIENVENID@ A HANGMAN! ")
 
    nombreUsuario = input("Ingrese su nombre: ")

    difficulty = '' #empieza vacio
    while True:
        print("ingrese nivel de dificultad: L - libre, F - Facil, M - Medio, D - Dificil.") #se ponen los 4 niveles para
        difficulty = input().upper()                                                        #luego permitir el ingreso del
                                                                                            #nivel  

        if difficulty in ["L", "F", "M","D"]:# se pone en lista para que solo acepte letras solas

            print("ingrese la cantidad de letras de la palabra EN NUMEROS: ")
            cantidad_letras_usuario = input() 
            cantidad_letras_usuario = int(cantidad_letras_usuario)
            print( cantidad_letras_usuario )
            # se le pide al user poner un numero y se hace entero
            

    
            break
            

        else:
            print("Ingrese un valor valido")
            
    clear_output() 
    sleep(0.5)
        
    return cantidad_letras_usuario,difficulty

def listaElegida(cantidad_letras_usuario,lista_palabras):
    
    lista_elegida = [] # esto es para sacar de la 'bolsa de palabras grandes' la bolsa del largo de la palabra elegida

    for elemento in lista_palabras:

        if len(elemento) == cantidad_letras_usuario: # si el largo de un elemento en la bolsa es = a 
                                                    #la cantidad que el user puso
            lista_elegida.append(elemento)

        else:
            continue
                  
    word = random.choice(lista_elegida)
    word = word.upper()  
    
    return word
                
def juego(word,difficulty,fases):
    
    word_espacios =  list(len(word)*"_") # se hace una lista con la cantidad de campos de la palabra
    palabra_adivinada = False 
    tries = 0 #empieza en 0 y va cambiando
    letras_adivinadas = []
    palabras_adivinadas = []

    # se ponen los valores correspondientes por nivel 
    if difficulty == "L":
        tries += 1
        
    elif difficulty == "F":
        tries = 12
        
    elif difficulty == "M" and tries <= 8:
        tries = 8
    
            
    elif difficulty == "D" and tries <= 5:
        tries = 5
        fases[tries+2]
           
             
    
    while not palabra_adivinada and tries >= 0 :
    
        adivinado = input("adivina una letra o palabra: ").upper() # aca se guardan las letras o palabras que la persona va poniendo

        #esto es para las letras individuales, por eso == 1(largo)
        if len(adivinado) == 1 and adivinado.isalpha(): #isalpha es para ver si es una letra efectivamente

            if adivinado in letras_adivinadas: 
                clear_output()                                                      
                sleep(0.2)       
    
                print("ya intentaste esa letra: " ,adivinado )
            

            elif adivinado not in word and difficulty == "L": # se divide en un condicional aparte para poder 
                
                clear_output()                                                      
                sleep(0.2)       
                print(adivinado, "no esta en la palabra ")   #sumar en vez de restar intentos. 

                print("intentos: ", tries)
                tries+=1
                letras_adivinadas.append(adivinado)
                 
                                        
            elif adivinado not in word:
                clear_output()                                                      
                sleep(0.2)       
    
                print(adivinado, "no esta en la palabra ")

                print("intentos: ", tries)
                tries-=1
                letras_adivinadas.append(adivinado)
                
                if difficulty == "F": #Aca se determina como se va mostrando el hangman dependiendo del nivel.   
                    if tries >= 8:    # se queda en una imagen hasta que cumpla la cantidad de valores necesarios para
                        print(fases[9]) #que se continuen mostrando las siguientes imagenes. 
                    else:
                        print(fases[tries])
                        
                 
                elif difficulty == "M":
                    if tries >= 6:
                        print(fases[9])
                    else:
                        print(fases[tries+2])
                        
                elif difficulty == "D":
                    print(fases[tries+2])
                 
                                
            else:
                clear_output()                                                      
                sleep(0.2)       
    
                print("Bien,", adivinado, "esta en la palabra!")
                letras_adivinadas.append(adivinado)
                lista_word = list(word_espacios) #lista word es para hacer la lista pasada

                indices = [ind for ind, word in enumerate(word) if word == adivinado] #el enumerate 
                                                                                    #devuelve un ind(indice) y su valor(word)
                                                                                    #esto es para ver en que 
                                                                                    #lugar esta la letra
                                                                            
                for index in indices: #aca para directamente por cada indice en el grupo de indices 
                    lista_word[index] = adivinado
                    word_espacios = "".join(lista_word) 

                if "_" not in word_espacios:
                    palabra_adivinada = True #cierra el ciclo
                    
    

        #palaras completas    
        elif len(adivinado) == len(adivinado) and adivinado.isalpha():
                if adivinado in palabras_adivinadas:
                    print("ya intentaste esa letra: ",adivinado )
                     

                elif adivinado != word and difficulty == "L": # se divide en un condicional aparte para poder 
                    print(adivinado, "no es la palabra!! ")   #sumar en vez de restar intentos. 

                    print("intentos: ", tries)
                    tries+=1
                    letras_adivinadas.append(adivinado)

                elif adivinado != word:
                    print(adivinado, " no es la palabra!!")
                    print("intentos: ", tries)
                    tries -= 1
                    letras_adivinadas.append(adivinado) 
                    
                    if difficulty == "F":   
                        if tries >= 8:    
                            print(fases[9])
                        else:
                            print(fases[tries])
                        
                 
                    elif difficulty == "M":
                        if tries >= 6:
                            print(fases[9])
                        else:
                            print(fases[tries+2])

                    elif difficulty == "D":
                        print(fases[tries+2])
                 
                        
            
                else:
                    palabra_adivinada = True
                    word_espacios = word     
                

        else:
            print("valor invalido")
            
     
        print(word_espacios)
        print("\n")


    if palabra_adivinada:
        print("Adivinaste!!!")
    else:
        print("Lastima, ya no tienes intentos. La palabra era " + word + ".")
        
    return adivinado

def hangman():
    fases = [ """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """
               , """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                  """
                   --------
                   |      |
                   |      O
                   |      |
                   |      
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """,
                """
                   --------
                   |      
                   |      
                   |    
                   |      
                   |     
                   -
                """
        
    ]
    return fases
  

def menu_principal():
    
    cantidad_letras_usuario,difficulty = datosUsuario(lista_palabras)
    word = listaElegida(cantidad_letras_usuario,lista_palabras)
    fases = hangman()
    juego(word,difficulty,fases)
    
    
    while input("Quieres Jugar De Nuevo? (S/N) ").upper() == "S":
        clear_output()                                                      
        sleep(0.2)
        
        cantidad_letras_usuario,difficulty = datosUsuario(lista_palabras)
        word = listaElegida(cantidad_letras_usuario,lista_palabras)
        fases = hangman()
        juego(word,difficulty,fases)
     
    print("=====================================================================")   
    print("Gracias por jugar :) ")
        
        
menu_principal()
````
